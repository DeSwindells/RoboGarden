```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from matplotlib.pyplot import figure
```

```python
# Read the dataset into the Pandas’ DataFrame!
df = pd.read_csv('data.csv')
df
```

```python
# Does the dataset include any missing values? 
df.isnull().sum()
```

```python
# If so, delete the missing values entries!
df.dropna()
df
```

```python
# Dropping Unnecessary Columns
# Choose the features you think are relevant to our analysis! There are a lot of features in this dataset, but we have to make our model’s training time reasonable for you.
# HINT: notice the fact that some of the data in the Breast Cancer dataset is irrelevant to the research such as the ID attribute.
print(df.columns)
df.drop(columns = ['id', 'radius_mean', 'radius_se', 'radius_worst', 'Unnamed: 32'], inplace = True)
df.columns
```

```python
# Perform any needed pre-processing on the chosen features, including:
# Scaling, Encoding, Dealing with NaN values, Normalization
# HINT: You need to output the result of your pre-processing to an output CSV called “data_refined.csv.”

#Scaling
scaler = StandardScaler()
df.iloc[:, 1:] = scaler.fit_transform(df.iloc[:, 1:])
#Encoding
encoder = LabelEncoder()
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
# NAN Values
nanValues = df.isnull().values.any()
print(nanValues)
# Save to new CSV
df.to_csv("data_refined.csv", index = False)
```
