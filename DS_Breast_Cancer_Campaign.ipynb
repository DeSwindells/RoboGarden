# Reading the Dataset

```python
# Load the previously preprocessed dataset saved in the CSV file “data_refined.csv” into a Pandas DataFrame.
import pandas as pd
df = pd.read_csv("C:/Users/dswinde/OneDrive - The University of Western Ontario/Documents/VSPY/RoboGarden/Module-2/data_refined.csv")
df
```

# Feature Selection

```python
# Choose only the most important features for training your classifier (choose the features with the highest correlation to the target data.)
# The label in this case is the ‘Diagnosed’ column which has 2 distinct variables: M/1: Malignant Tumor, B/0: Benign Tumor
# Calculate the correlation of all the features to their target labels.
df_target = df[['diagnosis']]
df_features = df.drop('diagnosis', axis = 1)
df_correlations = df_features.apply(lambda x: x.corr(df['diagnosis']))
df_correlations
```

```python
#  Choose the most correlated features above a certain limit for training, Output a list of important feature names.
df_relevant = df_correlations[abs(df_correlations>0.6)] # 0.58 to add in: compactness_mean and compactness_worst for a larger sample set

df2 = df[['perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean', 'perimeter_worst', 'area_worst', 'concavity_worst', 'concave points_worst']]
df_relevant
```
