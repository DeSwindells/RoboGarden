# Feature Detection

```python
# Extract the following features:
# Gender, Age, Scholarship, Hipertension, Diabetes, Alcoholism, Handcap, SMS_received
# Note: you may see that some datasets names are not written using proper spelling. It is related to the dataset, for example: Hipertension=Hypertension, Handcap=Handicap
print(df.dtypes)
extracted = df[['Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']]
target = df[['No-show']]
```

# Preproccessing

```python
# Perform any needed preprocessing of the chosen features, including: Scaling, Encoding, Dealing with NaN values
# Hint: use only the preprocessing steps you believe are useful.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
extracted[['Age']] = scaler.fit_transform(extracted[['Age']])
#extracted.iloc[:, 0] = scaler.fit_transform(extracted.iloc[:,0])

#target = pd.get_dummies(target, dtype=int)
#extracted_df = pd.concat([extracted, target], axis = 1)
# nan values
print("Are the any NAN Values?:",extracted.isnull().values.any())

extracted
```

# Splitting The Data

```python
# Split your data as follows: 80% training set, 10% validation set, 10% test set
from sklearn.model_selection import train_test_split
encoded_y = target.values
encoded_X = extracted.iloc[:, 0:]
X_train, X_test, y_train, y_test = train_test_split(encoded_X, encoded_y, test_size=0.1, train_size=0.8, random_state=0)
X_train, X_validate, y_train, y_validate = train_test_split(encoded_X, encoded_y, test_size=0.1,train_size=0.8, random_state=0)
```

# Training Tree-based Classifiers

```python
# Use a decision tree classifier model to train your data.
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini', splitter='best')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)

#Print the accuracy score of your final classifier.
score = model.score(X_test,y_test)
print(score)
```

```python
# Print the confusion matrix.
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_test, y_pred)
print(conf)
```
