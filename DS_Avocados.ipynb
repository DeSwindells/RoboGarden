# Read the Dataset

```python
# Read the dataset into a Pandas Dataframe.
import pandas as pd
df = pd.read_csv("C:/Users/dswinde/OneDrive - The University of Western Ontario/Documents/VSPY/RoboGarden/Module-4/avocado.csv")
df
```

# Extract Features
Exclude the region and date from the considered features.

```python
df_features = df.drop(['Unnamed: 0', 'Date', 'region'], axis = 1)
df_features
```

# Perform Preprocessing
Perform any needed pre-processing on the chosen features including: Scaling; Encoding; and Dealing with Nan values.
Hint: Use only the preprocessing steps for this mini-project.

```python
# Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_features.iloc[:, 1:9] = scaler.fit_transform(df_features.iloc[:, 1:9])
df_features[['year']]= scaler.fit_transform(df_features[['year']])
df_features
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit_transform(df_features['type'])
df_features['type']= encoder.transform(df_features['type'])
df_features
```

```python
#Nan Values
print("Are the any NAN Values?:",df_features.isnull().values.any())
```

# Split the Data

```python
# 80% training set, 10% validation set, 10% test set
from sklearn.model_selection import train_test_split
encoded_y = df_features.iloc[:,0].values
print(encoded_y)
encoded_X = df_features.iloc[:, 1:]
print(encoded_X)
X_train, X_test, y_train, y_test = train_test_split(encoded_X, encoded_y, test_size=0.1, train_size=0.2, random_state=0) # split the data to a test set by 20% using 0.2
X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size=0.1,train_size=0.5, random_state=0) # split the test set by %50 using 0.5 
```

# Train KNN Regression
Use a KNN regressor model to train your data. Choose the best k for the KNN algorithm by trying different values and validating performance on the validation set.

```python
from sklearn.neighbors import KNeighborsRegressor 
# Choose the best value of K 
scores = []
results = 0
best_score = 0
neighbors = range(1,20)

for i in neighbors:
    model = KNeighborsRegressor(n_neighbors=i).fit(X_train, y_train)
    results = model.score(X_test, y_test)
    scores.append(round(results,2))
    if results > best_score:
        best_score = results
        best_k = i
        best_model = model
print(scores)
print(best_k)
```

```python
# Visualize KNN Scores
import matplotlib.pyplot as plt
plt.plot(neighbors, scores)
```

```python
# Print the R-Squared score of your final K-Nearest Neighbor (KNN) regressor
accuracy = best_model.score(X_validate, y_validate)
print('The Best model has an accuracy of: ', round(accuracy,2))
```
