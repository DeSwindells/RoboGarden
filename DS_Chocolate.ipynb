```python
import pandas as pd
df = pd.read_csv("flavors_of_cacao.csv")
df
```

```python
df.columns = ['Company', 'SpecificBeanOrigin',
       'REF', 'ReviewDate', 'CocoaPercent', 'CompanyLocation', 'Rating',
       'BeanType', 'BroadBeanOrigin']
```

```python
# Does the dataset include any missing values? If so, delete the missing values entries!
df.dropna()
```

```python
# Visualize the rating column with a histogram!
import matplotlib.pyplot as plt
ratings = df.Rating.values
plt.hist(ratings)
plt.xlabel("Ratings")
```

```python
# Change the type of values in the column percent from string values to numerical values. **ONLY RUN THIS BLOCK ONCE OR IT WILL NOT WORK AGAIN, YOU WILL HAVE TO RUN ALL PREVIOUS BLOCKS TO FIX
#df['CocoaPercent'] = df['CocoaPercent'].str.replace(r'\D', "").astype(int) ---- This is the example given by RoboGarden, it does not work
df['CocoaPercent'] = df['CocoaPercent'].str.replace(r'\D', "", regex = True).astype(int) # ****Example given on RoboGarden doesn not work without adding 'regex = True'
```

```python
# Plot the converted numerical Cocoa Percent values against the Rating values!
cocoaP = df.CocoaPercent.values
plt.scatter(cocoaP, ratings)
plt.xlabel("CocoaPercent")
```

```python
# From what you see, does more cocoa in a bar correspond to a higher rating?
# Hint: Try a scatter plot with small alpha, e.g., 0.1, to flush out the density of each point.
plt.scatter(cocoaP, ratings, alpha = 0.1)
plt.xlabel("CocoaPercent")
```

```python
# Drop Outliers in Cocoa Percentage Data
df.drop(df[df.CocoaPercent > 100].index, inplace = True)
ratingsDrpd = df.Rating.values
percentage = df.CocoaPercent.values
plt.scatter(percentage, ratingsDrpd, alpha = 0.1)
plt.xlabel("CocoaPercent")
```
